```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

```python
acetone = (14.5463,2940.46,237.22)
acetonitrile = (14.2724,2945.224,224)
nitromethane = (14.2043,2972.64,209)
ethanol = (16.8958,3795.17,230.918)
water = (16.3872,3885.7,230.17)
methanol = (16.5785,3638.27,239.5)

def Antoine(T,A,B,C):
  return np.exp(A - B / (T + C))

def BP(T):
  species = nitromethane
  psat=Antoine(T,*species)
  error=psat-100
  return error
T1 = Antoine(25,*acetonitrile)
root = fsolve(BP,60)

Psat1 = Antoine(10,*acetone)
Psat2 = Antoine(10,*acetonitrile)
Psat3 = Antoine(10,*nitromethane)
Psat4 = Antoine(10,*ethanol)
Psat5 = Antoine(10,*water)


bubble = .2*Psat1+.2*Psat2+.2*Psat3+.2*Psat4+.2*Psat5
print("Pbubble =", bubble)
dew = 1/(.2/Psat1+.2/Psat2+.2/Psat3+.2/Psat4+.2/Psat5)
print("Pdew =" ,dew)
```

```python
def f(x):
  y=(x+10)**2+20
  return y
xx=np.linspace(-50,50,200)
yy=f(xx)
plt.plot(xx,yy)
plt.title("equation : (x+10)^2+20")
plt.show()
root = fsolve(f,10)
print('root =',root)
```

```python
Psat1 = Antoine(10,*acetone)
Psat2 = Antoine(10,*acetonitrile)
Psat3 = Antoine(10,*nitromethane)
Psat4 = Antoine(10,*ethanol)
Psat5 = Antoine(10,*water)
P=3.75

def V(x):
  k1=Psat1/P
  k2=Psat2/P
  k3=Psat3/P
  k4=Psat4/P
  k5=Psat5/P
  y = (.2*k1)/(1+x*(k1-1))+(.2*k2)/(1+x*(k2-1))+(.2*k3)/(1+x*(k3-1))+(.2*k4)/(1+x*(k4-1))+(.2*k5)/(1+x*(k5-1))-1
  return y
V = fsolve(V,.4)
print(V)
def L(x):
  k1=Psat1/P
  k2=Psat2/P
  k3=Psat3/P
  k4=Psat4/P
  k5=Psat5/P
  y = (.2*k1)/(k1+x*(1-k1))+(.2*k2)/(k2+x*(1-k2))+(.2*k3)/(k3+x*(1-k3))+(.2*k4)/(k4+x*(1-k4))+(.2*k5)/(k5+x*(1-k5))-1
  return y
L_x=fsolve(L,.4)
print(L_x)
```

```python
T=25
p1sat = Antoine(T,*ethanol)
p2sat = Antoine(T,*water)
for mol_frac in np.linspace(0,1,101):
  x1 = mol_frac
  x2 = 1 - x1 
  A = -1
  gamma1 = np.exp(A*x2**2)
  gamma2 = np.exp(A*x1**2)
  P_bubble = x1*gamma1*p1sat + x2*gamma2*p2sat  
  plt.plot(x1,P_bubble,'.r')
  y1 = mol_frac
  P_dew = 1 /(y1/(gamma1*p1sat) + (1-y1)/(gamma2*p2sat))
  plt.plot(y1,P_dew,'.k')
plt.xlabel('mole fraction')
plt.ylabel('Pressure (kPa)')
plt.title('PXY A=-1')
plt.show()
```
